We used Matplotlib to create a histogram â€” a special kind of graph that shows how many cereals fall into different calorie ranges.
We divided the range of calories into 20 bins (small groups), so we could see exactly where most cereals are.
The X-axis shows the number of calories.
The Y-axis shows how many cereals fall into each calorie group

```python
import matplotlib.pyplot as plt

df['calories'].hist(bins=20)
plt.title('Calories Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

We created a scatter plot (each point represents one cereal).
X-axis: Amount of sugar (per each cereal).
Y-axis: Rating (of each cereal).
Each point shows a cerealâ€™s sugar content and how was its rating 

The result of this visualization is  : The more sugar a cereal had, the lower its rating tended to be.

```python
plt.scatter(df['sugars'], df['rating'])
plt.xlabel('Sugars')
plt.ylabel('Rating')
plt.title('Sugars vs Rating')
plt.show()
```

We used groupby() to group cereals by manufacturer (like Kellogg's, General Mills, Quaker, etc.).
For each group, we calculated the mean (average) rating.
Then, we plotted this as a bar chart â€” where each bar represents a manufacturerâ€™s average cereal rating.

The result of this visualization:

Some brands like Kelloggâ€™s (K) and General Mills (G) stood tall â€” showing strong average ratings across multiple cereals.
Others had lower average ratings, possibly due to producing more sugary, less healthy cereals.

```python
df.groupby('mfr')['rating'].mean().plot(kind='bar')
plt.title('Average Rating by Manufacturer')
plt.ylabel('Rating')
plt.show()
```
