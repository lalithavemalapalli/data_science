### Nutritional Comparision of selected Cereals (sugars,calories,vitamins)

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 12))
a.set_index('name')[['calories', 'sugars', 'vitamins']].plot(kind='bar', ax=ax)

# Customize the plot
plt.title("Nutritional Comparison of Selected Cereals (Health-Conscious Focus)")
plt.ylabel("Grams per Serving")
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
```

##### After checking the calories and sugar contents in the cereals now the customer comes to the fiber conetent and the protien content . which  cereal is having the highest protien and high fibre content in it .
```python
a=df[(df['fiber'] > 5)  & (df['protein'] > 3)]
a
```

##### The cereal which contains the highest fibre and hight rotien content it should also contains high vitamins also in it 

```python
a=df[(df['fiber'] > 5)  & (df['protein'] > 3) & (df['vitamins'] > 20)]
a
```

```python
a[['fiber', 'protein', 'vitamins']].plot(kind='bar', stacked=True, figsize=(10, 10), colormap='Set3')
plt.title("Stacked Bar Plot of Nutritional Content per Cereal")
plt.xlabel("Cereal Name")
plt.ylabel("Amount (g or %DV)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title="Nutrient")
plt.tight_layout()
plt.show()
```

##### Now we need to see the protien count for each cereal and group them by names 

```python
protein_by_cereal = df.groupby('name')['protein'].sum().reset_index()
protein_by_cereal
```

```python
fibre_by_cereal = df.groupby('name')['fiber'].sum().reset_index()
fibre_by_cereal
```

```python
vitamin_by_cereal = df.groupby('name')['vitamins'].sum().reset_index()
vitamin_by_cereal
```

```python
import matplotlib.pyplot as plt

df['calories'].hist(bins=20)
plt.title='Calories Distribution'
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

##### Now we are concluding this by getting the best rated high protien , high fibre , high vitamin and less calories and less sugar content cereal 

```python
healthy_cereals = df[
    (df['protein'] >= 3) &
    (df['fiber'] >= 5) &
    (df['vitamins'] >= 25) &
    (df['calories'] <= 100) &
    (df['sugars'] <= 5)
]
healthy_cereals
```
