##### Food is the main reason for the obesity evel . We can easily tell obesity level based on the persons food intake habits .Now let us calculate the intake of the food of a person . the inatke of food depends on the main meals , vegetable consumption , and taking food in between the meals , and alcohol consumption and it mainly depends on the intake of high calorie food . If a person is taking good vegetable consumption and do not taking food between meals , and does not addicted to the alcohol consuption and do not take any high calorie food then he is a helthy person he is not an overweighted person . But if a person takes high calorie foods and snacks in between the meals and alcohol consumption then he is an overweighted person . 

```python
def compute_food_intake(row):
    score = 0

    # High-calorie food 
    if row['FAVC'] == 'yes':
        score += 1

    # Main meals 
    if row['NCP'] >= 3:
        score += 0.5
    elif row['NCP'] <= 1:
        score += 1  # Possibly bingeing

    # Vegetable consumption 
    if row['FCVC'] < 2:
        score += 1

    # Eating between meals
    if row['CAEC'] in ['Frequently', 'Always']:
        score += 1

    # Alcohol consumption
    if row['CALC'] == 'Frequently':
        score += 1

    return score
```

```python
df['Food_Intake_Score'] = df.apply(compute_food_intake, axis=1)
df
```

##### How the intake of food is important for the obesity levels like that only means of transportation , and involvement in the physical activities playes a major role . the person who is involved in the physical activities and the person who has walking and public transportation has an daily routine he will be slightly healthy person . 

```python
def activity_score(faf):
    if faf >= 3:
        return 0
    elif faf >= 1:
        return 1
    else:
        return 2

def transport_score(mtrans):
    if mtrans in ['Walking', 'Bike']:
        return 0
    elif mtrans == 'Public_Transportation':
        return 1
    else:
        return 2

def estimate_obesity_activity_based(faf, mtrans):
    total = activity_score(faf) + transport_score(mtrans)
    if total <= 1:
        return "Normal_Weight"
    elif total == 2:
        return "near_to_overweight"
    else:
        return "overweight"

df['Obesity_From_Activity'] = df.apply(lambda row: estimate_obesity_activity_based(row['FAF'], row['MTRANS']), axis=1)
df
```

##### The person who is good in intake of the food and who always involve in the physical activities and use the walking or public transporattion as their means of transport but he frequently involve the inatke of the smoking and alcohol consumption also no use in the taking of the healthy food . the intake of the alcohol and smoking also increase the obesity level . it also plays a major role .

```python
# calculating the intake of alcohol consumption 
def alcohol_score(calc):
    mapping = {'no': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3}
    return mapping.get(calc, 0)

# calculating the smoking levels of a person 
def smoking_score(smoke):
    return 1 if smoke == 'yes' else 0

# combining both 
def substance_intake(calc, smoke):
    return alcohol_score(calc) + smoking_score(smoke)

df['Alcohol_Intake_Score'] = df['CALC'].apply(alcohol_score)
df['Smoking_Intake_Score'] = df['SMOKE'].apply(smoking_score)
df['addiction_score'] = df.apply(lambda row: substance_intake(row['CALC'], row['SMOKE']), axis=1)
df
```
