- **Objective**: Implement a custom aggregation function for group-level analysis.
- **Instructions**:

1. Define a custom function that calculates the **Sales per Transaction** ratio for each group (i.e., total `Sales` divided by the number of transactions).

Solution :
Here the task is to calculate the sales per each transaction for each group means each store for this:
- first we need to create the function called as sales_per_transaction
- The in that we need to create the variable called num_of_sales ih this variable we are calculating the total numner of sales using the `sum()` function 
- then we need to create the variable called the num_of_transaction in this we are calculating the total numnber of transaction per each sale using the `sum()` function 
- then we will use the `apply()` function to apply this function to the dataframe 

```python
def sales_per_transaction(df):
    return df.groupby(df['Store']).apply(lambda x : x['Sales'].sum() / x['Sales'].count())
sales_per_transaction(df)
```
